# Problem Statement

Modern social media applications such as Instagram handle millions of users who continuously upload photos, interact with content, and consume personalized feeds.

The challenge is to design a system that can:

- Efficiently store user information
- Process uploaded images
- Generate feeds
- Support likes and comments
- Manage follower relationships
- Deliver notifications
- Scale for increasing traffic

SnapShare solves these challenges using Python, Streamlit, SQLite, and modular system design principles.

Python Implementation
Step 1: Import Required Libraries

```python
import sqlite3
import hashlib
from PIL import Image
from datetime import datetime

print("Libraries Imported Successfully")
```

Step 2: User Model

```python
class User:

    def __init__(self, username, email):
        self.username = username
        self.email = email

    def display(self):
        print("Username:", self.username)
        print("Email:", self.email)


user = User("ragini", "ragini@gmail.com")
user.display()
```

Step 3: Password Hashing

```python
def hash_password(password):

    return hashlib.sha256(password.encode()).hexdigest()


password = "ragini123"

hashed_password = hash_password(password)

print("Original Password:", password)
print("Hashed Password:", hashed_password)
```

Step 4: Database Connection

```python
connection = sqlite3.connect("snapshare.db")

cursor = connection.cursor()

print("Database Connected Successfully")
```

Step 5: Users Table Creation

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS users(
id INTEGER PRIMARY KEY,
username TEXT,
email TEXT,
password TEXT
)
""")

connection.commit()

print("Users Table Created")
```

Step 6: Post Model

```python
class Post:

    def __init__(self, caption):
        self.caption = caption
        self.created_at = datetime.now()

    def show(self):
        print(self.caption)
        print(self.created_at)


post = Post("My First SnapShare Post")
post.show()
```

Step 7: Image Compression Logic

```python
def compress_image(input_file, output_file):

    image = Image.open(input_file)

    image.save(
        output_file,
        "JPEG",
        quality=75
    )

    print("Image Compressed Successfully")


print("Compression Module Ready")
```

Step 8: Feed Ranking Algorithm

```python
posts = [

    {"likes": 10, "comments": 5},
    {"likes": 20, "comments": 2},
    {"likes": 15, "comments": 8}

]

for post in posts:

    score = post["likes"] + (post["comments"] * 2)

    print("Trending Score:", score)
```

Step 9: Follow System

```python
followers = []

followers.append("user1")
followers.append("user2")

print("Followers Count:", len(followers))
```
